```python
import csv
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
area_death={}
areas_death={}
area_comfirmed={}
areas_comfirmed={}
area_recovered={}
areas_recovered={}
for num in range(1,22):
    filename='C:\\Users\\86188\\Desktop\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_daily_reports\\03-'+str(num).zfill(2)+'-2020.csv'
    with open(filename) as f:
        reader = csv.reader(f)
        header_row = next(reader)
        for row in reader:
            area_death[row[1]]=0
            area_comfirmed[row[1]]=0
            area_recovered[row[1]]=0
for num in range(1,22):
    filename='C:\\Users\\86188\\Desktop\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_daily_reports\\03-'+str(num).zfill(2)+'-2020.csv' 
    with open(filename) as f:
        reader1 = csv.reader(f)
        for row in reader1:
            if row[4]!=''and row[4]!='Deaths' and row[0]!=''and row[5]!='':
                area_death[row[1]]=area_death[row[1]]+int(row[4])
                area_comfirmed[row[1]]=area_comfirmed[row[1]]+int(row[3])
                area_recovered[row[1]]=area_recovered[row[1]]+int(row[5])
print(area_death)
#print( sorted(area_death.items(), key=lambda x: x[1], reverse=True)[0][0])
for i in range(0,5):
    areas_death[sorted(area_death.items(), key=lambda x: x[1], reverse=True)[i][0]]=int(sorted(area_death.items(), key=lambda x: x[1], reverse=True)[i][1])
    areas_comfirmed[sorted(area_comfirmed.items(), key=lambda x: x[1], reverse=True)[i][0]]=int(sorted(area_comfirmed.items(), key=lambda x: x[1], reverse=True)[i][1])
    areas_recovered[sorted(area_recovered.items(), key=lambda x: x[1], reverse=True)[i][0]]=int(sorted(area_recovered.items(), key=lambda x: x[1], reverse=True)[i][1])
print(areas_death)
print(areas_comfirmed)
print(areas_recovered)
areas1=areas_death.keys()
death=areas_death.values()
areas2=areas_comfirmed.keys()
comfirmed = areas_comfirmed.values()
areas3=areas_recovered.keys()
recovered = areas_recovered.values()
```

列举三月前五的death地区或国家(部分数据)

```python
plt.figure(dpi=128,figsize=(20,10))
plt.bar(areas1, death, label='death')
plt.title('death',fontsize=25)
plt.legend()
plt.xlabel('data',fontsize = 16)
plt.ylabel('top five num of death',fontsize = 24)
plt.show()
```

列举三月前五的comfirmed地区或国家(部分数据)

```python
plt.figure(dpi=128,figsize=(20,10))
plt.bar(areas2, comfirmed, label='comfirmed')
plt.title('comfirmed',fontsize=25)
plt.legend()
plt.xlabel('data',fontsize = 16)
plt.ylabel('top five num of comfirmed',fontsize = 24)
plt.show()
```

列举三月前五的recovered地区或国家(部分数据)

```python
plt.figure(dpi=128,figsize=(20,10))
plt.bar(areas3, recovered, label='recovered')
plt.title('recovered',fontsize=25)
plt.legend()
plt.xlabel('data',fontsize = 16)
plt.ylabel('top five num of recovered',fontsize = 24)
plt.show()
```

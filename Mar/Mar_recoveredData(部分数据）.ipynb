```python
import csv
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
time_recovered={}
time = []
for num in range(1,22):
    filename='C:\\Users\\86188\\Desktop\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_daily_reports\\03-'+str(num).zfill(2)+'-2020.csv'
    with open(filename) as f:
        reader = csv.reader(f)
        header_row = next(reader)
        for row in reader:
            time_recovered['03-'+str(num).zfill(2)]=0
for num in range(1,22):
    filename='C:\\Users\\86188\\Desktop\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_daily_reports\\03-'+str(num).zfill(2)+'-2020.csv' 
    with open(filename) as f:
        reader1 = csv.reader(f)
        for row in reader1:
            if row[5]!=''and row[5]!='Recovered':
                time_recovered['03-'+str(num).zfill(2)]=float(row[5])+time_recovered['03-'+str(num).zfill(2)]
print(time_recovered)
times,recovered=time_recovered.keys(),time_recovered.values()
for i in times:
    time.append(i)
```

```python
plt.figure(dpi=800)
plt.title('cumulative Recovered')
plt.xlabel("date",fontsize=14)
plt.ylabel("num of recovered",fontsize=14)
plt.tick_params(axis='both',which='major',labelsize=3)
plt.scatter(time,recovered,s=100)
plt.show()
```

```python
new_time_recovered={}
for key in time_recovered.keys():
    if key!='03-01':
        new_time_recovered[key]=time_recovered[key]-time_recovered['03-'+str(int(key[3:5])-1).zfill(2)]
print(new_time_recovered)
time1=new_time_recovered.keys()
recover1=new_time_recovered.values()
plt.figure(dpi=800)
plt.title('new recovered')
plt.xlabel("date",fontsize=14)
plt.ylabel("num of recovered",fontsize=14)
plt.tick_params(axis='both',which='major',labelsize=3)
plt.scatter(time1,recover1,s=100)
plt.show()
```

```python
recover=[]
for i in recovered:
    recover.append(i)
fig = plt.figure(dpi=128,figsize=(10,10))
plt.title('cumulative Recovered',fontsize=24)
plt.xlabel('date',fontsize=16)
plt.ylabel('num of recovered',fontsize=16)
plt.plot(time,recover,c='red')
plt.tick_params(axis='both',which='major',labelsize=5)
plt.show()
```

```python
recover2=[]
for i in recover1:
    recover2.append(i)
time2=[]
for i in time1:
    time2.append(i)
fig = plt.figure(dpi=128,figsize=(10,10))
plt.title('New recovered',fontsize=24)
plt.xlabel('date',fontsize=16)
plt.ylabel('num of recovered',fontsize=16)
plt.plot(time2,recover2,c='red')
plt.tick_params(axis='both',which='major',labelsize=5)
plt.show()
```

```python
plt.figure(dpi=128,figsize=(18,10))
plt.bar(time, recover, label='recovered')
plt.title('cumulative recovered',fontsize=25)
plt.legend()
plt.xlabel('data',fontsize = 16)
plt.ylabel('num of recovered',fontsize = 24)
plt.show()
```

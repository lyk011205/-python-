```python
import csv
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
time_recovered={}
time_death={}
r_d=[]
time = []
for num in range(1,30):
    filename='C:\\Users\\86188\\Desktop\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_daily_reports\\02-'+str(num).zfill(2)+'-2020.csv'
    with open(filename) as f:
        reader = csv.reader(f)
        header_row = next(reader)
        for row in reader:
            time_recovered['02-'+str(num).zfill(2)]=0
            time_death['02-'+str(num).zfill(2)]=0
for num in range(1,30):
    filename='C:\\Users\\86188\\Desktop\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_daily_reports\\02-'+str(num).zfill(2)+'-2020.csv' 
    with open(filename) as f:
        reader1 = csv.reader(f)
        for row in reader1:
            if row[5]!=''and row[5]!='Recovered'and row[4]!='Deaths'and row[4]!='':
                time_recovered['02-'+str(num).zfill(2)]=float(row[5])+time_recovered['02-'+str(num).zfill(2)]
                time_death['02-'+str(num).zfill(2)]=float(row[4])+time_death['02-'+str(num).zfill(2)]
times,recovered=time_recovered.keys(),time_recovered.values()
times,deaths=time_death.keys(),time_death.values()
for i in times:
    time.append(i)
death=[]
for i in deaths:
    death.append(i)
recover=[]
for i in recovered:
    recover.append(i)
lens = len(time)
for i in range(lens):
    r_d.append(recover[i]-death[i])
print(r_d)
print(time)
```

```python
plt.figure(dpi=800)
plt.title('cumulative Recovered - Death')
plt.xlabel("date",fontsize=14)
plt.ylabel("num of recovered - death",fontsize=14)
plt.tick_params(axis='both',which='major',labelsize=3)
plt.scatter(time,r_d,s=100)
```

```python
new={}
for key in time_recovered.keys():
    if key!='02-01':
        new[key]=time_recovered[key]-time_recovered['02-'+str(int(key[3:5])-1).zfill(2)]-time_death[key]+time_death['02-'+str(int(key[3:5])-1).zfill(2)]
print(new)
time1=new.keys()
value1=new.values()
plt.figure(dpi=800)
plt.title('new recovered-death')
plt.xlabel("date",fontsize=14)
plt.ylabel("num of  recovered-death",fontsize=14)
plt.tick_params(axis='both',which='major',labelsize=3)
plt.scatter(time1,value1,s=100)
plt.show()
```

```python
fig = plt.figure(dpi=128,figsize=(10,6))
plt.title('cumulative Recovered - Death',fontsize=24)
plt.xlabel('date',fontsize=16)
plt.ylabel('num of recovered - death',fontsize=16)
plt.plot(time,r_d,c='red')
plt.tick_params(axis='both',which='major',labelsize=5)
plt.show()
```

```python
value2=[]
for i in value1:
    value2.append(i)
time2=[]
for i in time1:
    time2.append(i)
fig = plt.figure(dpi=128,figsize=(10,10))
plt.title('New recovered - death',fontsize=24)
plt.xlabel('date',fontsize=16)
plt.ylabel('num of recovered - death',fontsize=16)
plt.plot(time2,value2,c='red')
plt.tick_params(axis='both',which='major',labelsize=5)
plt.show()
```

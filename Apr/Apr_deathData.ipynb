```python
import csv
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
time_death={}
time = []
for num in range(1,31):
    filename='C:\\Users\\86188\\Desktop\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_daily_reports\\04-'+str(num).zfill(2)+'-2020.csv'
    with open(filename) as f:
        reader = csv.reader(f)
        header_row = next(reader)
        for row in reader:
            time_death['04-'+str(num).zfill(2)]=0
for num in range(1,31):
    filename='C:\\Users\\86188\\Desktop\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_daily_reports\\04-'+str(num).zfill(2)+'-2020.csv' 
    with open(filename) as f:
        reader1 = csv.reader(f)
        for row in reader1:
            if row[8]!=''and row[8]!='Deaths':
                time_death['04-'+str(num).zfill(2)]=float(row[8])+time_death['04-'+str(num).zfill(2)]
print(time_death)
times,deaths=time_death.keys(),time_death.values()
for i in times:
    time.append(i)
```

```python
plt.figure(dpi=800)
plt.title('cumulative death')
plt.xlabel("date",fontsize=14)
plt.ylabel("num of death",fontsize=14)
plt.tick_params(axis='both',which='major',labelsize=3)
plt.scatter(time,deaths,s=100)
plt.show()
```

```python
new_time_death={}
for key in time_death.keys():
    if key!='04-01':
        new_time_death[key]=time_death[key]-time_death['04-'+str(int(key[3:5])-1).zfill(2)]
print(new_time_death)
time1=new_time_death.keys()
death1=new_time_death.values()
plt.figure(dpi=800)
plt.title('new death')
plt.xlabel("date",fontsize=14)
plt.ylabel("num of death",fontsize=14)
plt.tick_params(axis='both',which='major',labelsize=3)
plt.scatter(time1,death1,s=100)
plt.show()
```

```python
death=[]
for i in deaths:
    death.append(i)
fig = plt.figure(dpi=128,figsize=(10,10))
plt.title('cumulative death',fontsize=24)
plt.xlabel('date',fontsize=16)
plt.ylabel('num of death',fontsize=16)
plt.plot(time,death,c='red')
plt.tick_params(axis='both',which='major',labelsize=5)
plt.show()
```

```python
death2=[]
for i in death1:
    death2.append(i)
time2=[]
for i in time1:
    time2.append(i)
fig = plt.figure(dpi=128,figsize=(10,10))
plt.title('New death',fontsize=24)
plt.xlabel('date',fontsize=16)
plt.ylabel('num of death',fontsize=16)
plt.plot(time2,death2,c='red')
plt.tick_params(axis='both',which='major',labelsize=5)
plt.show()
```

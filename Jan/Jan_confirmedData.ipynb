```python
import csv
```

```python
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
time_confirmed={}
time = []
for num in range(22,32):
    filename='C:\\Users\\86188\\Desktop\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_daily_reports\\01-'+str(num).zfill(2)+'-2020.csv'
    with open(filename) as f:
        reader = csv.reader(f)
        header_row = next(reader)
        for row in reader:
            time_confirmed['01-'+str(num).zfill(2)]=0
for num in range(22,32):
    filename='C:\\Users\\86188\\Desktop\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_daily_reports\\01-'+str(num).zfill(2)+'-2020.csv' 
    with open(filename) as f:
        reader1 = csv.reader(f)
        for row in reader1:
            if row[3]!=''and row[3]!='Confirmed':
                time_confirmed['01-'+str(num).zfill(2)]=float(row[3])+time_confirmed['01-'+str(num).zfill(2)]
print(time_confirmed)
times,confirmed=time_confirmed.keys(),time_confirmed.values()
for i in times:
    time.append(i)
```

```python
plt.figure(dpi=800)
plt.title('cumulative Confirmed')
plt.xlabel("date",fontsize=14)
plt.ylabel("num of confirmed",fontsize=14)
plt.tick_params(axis='both',which='major',labelsize=3)
plt.scatter(time,confirmed,s=100)
plt.show()
```

```python
new_time_confirmed={}
for key in time_confirmed.keys():
    if key!='01-22':
        new_time_confirmed[key]=time_confirmed[key]-time_confirmed['01-'+str(int(key[3:5])-1).zfill(2)]
print(new_time_confirmed)
time1=new_time_confirmed.keys()
confirmed1=new_time_confirmed.values()
plt.figure(dpi=800)
plt.title('new confirmed')
plt.xlabel("date",fontsize=14)
plt.ylabel("num of confirmed",fontsize=14)
plt.tick_params(axis='both',which='major',labelsize=3)
plt.scatter(time1,confirmed1,s=100)
plt.show()
```

```python
confirm=[]
for i in confirmed:
    confirm.append(i)
fig = plt.figure(dpi=128,figsize=(10,10))
plt.title('cumulative Confirmed',fontsize=24)
plt.xlabel('date',fontsize=16)
plt.ylabel('num of confirmed',fontsize=16)
plt.plot(time,confirm,c='red')
plt.tick_params(axis='both',which='major',labelsize=5)
plt.show()
```

```python
confirmed2=[]
for i in confirmed1:
    confirmed2.append(i)
time2=[]
for i in time1:
    time2.append(i)
fig = plt.figure(dpi=128,figsize=(10,10))
plt.title('New confirmed',fontsize=24)
plt.xlabel('date',fontsize=16)
plt.ylabel('num of confirmed',fontsize=16)
plt.plot(time2,confirmed2,c='red')
plt.tick_params(axis='both',which='major',labelsize=5)
plt.show()
```
